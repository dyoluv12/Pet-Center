```python
df_bus_station = pd.read_csv('서울시 정류장마스터 정보.csv', encoding="CP949")
df_bus_station
```

```python
df_bus_station['정류장_ID'] = df_bus_station['정류장_ID'].astype(str)
yangcheon_bus_station_data = df_bus_station[df_bus_station['정류장_ID'].str.startswith('114')]
yangcheon_bus_station_data
 #양천구 정류장분류 ID는 144로 시작
```

```python
yangcheon_bus_station = yangcheon_bus_station_data.reset_index(drop=True).drop(columns=['버스도착정보안내기_설치_여부', '정류장_유형', '정류장_번호'])
yangcheon_bus_station #필요없는 열 삭제
```

```python
yangcheon_bus_station.to_csv('서울시 양천구 버스정류장.csv', index=False)
```
